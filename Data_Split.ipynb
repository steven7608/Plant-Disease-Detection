```python
import os
import shutil
import random
```

```python
root_dir = "/content/drive/MyDrive/Colab Notebooks/Plant Disease Dataset/Dataset"
train_dir = "/content/drive/MyDrive/Colab Notebooks/Plant Disease Dataset/Train"
val_dir   = "/content/drive/MyDrive/Colab Notebooks/Plant Disease Dataset/Validation"
```

```python
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
```

```python
folders = [f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, f))]

print("Found", len(folders), "class folders:\n")
for folder in folders:
    print(folder)
```

```python
classes = os.listdir(root_dir)

for cls in classes:
    cls_path = os.path.join(root_dir, cls)
    if not os.path.isdir(cls_path):
        continue

    print(f"Processing class: {cls}")

    # Create class folder in train/val
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(val_dir, cls), exist_ok=True)

    # List and shuffle file names
    files = os.listdir(cls_path)
    random.shuffle(files)

    # Compute 80/20 split
    split_idx = int(0.8 * len(files))
    train_files = files[:split_idx]
    val_files = files[split_idx:]

    # Copy files
    for f in train_files:
        shutil.copy(os.path.join(cls_path, f),
                    os.path.join(train_dir, cls, f))

    for f in val_files:
        shutil.copy(os.path.join(cls_path, f),
                    os.path.join(val_dir, cls, f))

print("Split into 80% train and 20% validation.")
```

```python
folders = [f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(train_dir, f))]

print("Found", len(folders), "class folders:\n")
for folder in folders:
    print(folder)
```

```python
folder_info = []

for cls in os.listdir(root_dir):
    cls_path = os.path.join(root_dir, cls)
    if not os.path.isdir(cls_path):
        continue

    # Count only actual images
    image_files = [
        f for f in os.listdir(cls_path)
        if f.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    folder_info.append((cls, len(image_files)))

# Sort by image count
folder_info.sort(key=lambda x: x[1])

print("RESULTS:\n")
for name, count in folder_info:
    print(f"{name:50s}  ->  {count} images")

print("\nSUMMARY:")
print(f"Total folders scanned: {len(folder_info)}")
print(f"Folders with images:   {sum(1 for _,c in folder_info if c > 0)}")
print(f"Folders with 0 images: {sum(1 for _,c in folder_info if c == 0)}")
```

```python
folders = [f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(val_dir, f))]

print("Found", len(folders), "class folders:\n")
for folder in folders:
    print(folder)
```

```python
classes = [c for c in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, c))]

for cls in classes:
    cls_path = os.path.join(root_dir, cls)

    # Create destination folders safely
    train_cls_dir = os.path.join(train_dir, cls)
    val_cls_dir   = os.path.join(val_dir, cls)

    os.makedirs(train_cls_dir, exist_ok=True)
    os.makedirs(val_cls_dir, exist_ok=True)

    # Verify creation
    if not os.path.exists(train_cls_dir):
        print(f"ERROR: Could not create {train_cls_dir}")
        continue

    # List images
    files = [f for f in os.listdir(cls_path)
             if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    random.shuffle(files)

    split_idx = int(0.8 * len(files))
    train_files = files[:split_idx]
    val_files   = files[split_idx:]

    # Copy files
    for f in train_files:
        try:
            shutil.copy(os.path.join(cls_path, f),
                        os.path.join(train_cls_dir, f))
        except Exception as e:
            print("Copy error:", e)

    for f in val_files:
        try:
            shutil.copy(os.path.join(cls_path, f),
                        os.path.join(val_cls_dir, f))
        except Exception as e:
            print("Copy error:", e)

print("Done splitting.")
```
